```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture("dataset/video_1.avi")
frames = []
while True:
    # get a frame
    ret, frame = cap.read()
    if ret:
        frames.append(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))
    else:
        break
frames = np.array(frames)
```

### results
order|k|alpha
-|-|-
1|3|1e-2
2|3|1e-1

```python
k = 3
threshold = 2.5
alpha = 1e-2

weights = np.random.uniform(0, 1, size=(list(frames.shape)[1:] + [k]))
weights = weights / np.expand_dims(np.sum(weights, axis=2), axis=2) # uniformization
mus = np.random.uniform(0, 255.0, size=(list(frames.shape)[1:] + [k]))
rhos = np.random.uniform(5.0, 6.0, size=(list(frames.shape)[1:] + [k]))

masks = np.ones_like(frames) * 255
gaussian = lambda _X, _mu, _sigma: np.exp(-(_X - _mu) ** 2 /(0.5 *  _sigma)) / ((2 * np.pi) ** 0.5 * _sigma)
```

```python
for frame in range(frames.shape[0]):
    rank = np.argsort(weights / rhos, axis=2) # from low to high
    for i in range(frames.shape[1]):
        for j in range(frames.shape[2]):
            counter = 0
            for index in reversed(range(k)):
                #if gaussian(frames[frame, i, j], mus[i, j, rank[i, j, index]], rhos[i, j, rank[i, j, index]]) > threshold:
                if np.abs(frames[frame, i, j] - mus[i, j, rank[i, j, index]]) < threshold * rhos[i, j, rank[i, j, index]]:
                    weights[i, j, rank[i, j, index]] = (1 - alpha) * weights[i, j, rank[i, j, index]] + alpha
                    for index_ in reversed(range(index)): # normalization for other weights
                        weights[i, j, index_] = (1 - alpha) * weights[i, j, index_]
                    #beta = alpha / weights[i, j, rank[i, j, index]]
                    beta = alpha * gaussian(frames[frame, i, j], mus[i, j, rank[i, j, index]], rhos[i, j, rank[i, j, index]])
                    mus[i, j, rank[i, j, index]] = (1 - beta) * mus[i, j, rank[i, j, index]] + beta * frames[frame, i, j]
                    rhos[i, j, rank[i, j, index]] = np.sqrt((1 - beta) * rhos[i, j, rank[i, j, index]] ** 2 + beta * (frames[frame, i, j] - rhos[i, j, rank[i, j, index]]) ** 2)
                    masks[frame, i, j] = 0
                    continue
                else:
                    weights[i, j, rank[i, j, index]] = (1 - alpha) * weights[i, j, rank[i, j, index]]
                    counter += 1
            if counter == 3:
                mus[i, j, rank[i, j, 0]] = frames[frame, i, j]
                rhos[i, j, rank[i, j, 0]] = np.random.uniform(10.0, 12.0)
                weights[i, j, rank[i, j, 0]] = np.random.uniform(0.0, 0.2)
                weights[i, j] = weights[i, j] / np.sum(weights[i, j])
    print(str(frame) + ":" + str(np.sum(masks[frame])))
```
